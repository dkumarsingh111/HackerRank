```python
def frac(n):
    # TODO: Implement the recursive factorial function
    if n <= 1:
        return 1
    return n * frac(n - 1)
```

```python
def series(ip):
    # TODO: Implement the logic for computing the series
    if ip < 0:
        return 999.0

    total = 0.0
    for k in range(1, ip + 1):
        term = ((-1)**(k+1)) * ((frac(k)**2) / (frac(k) + k))
        total += term

    return round(total, 2)
```
